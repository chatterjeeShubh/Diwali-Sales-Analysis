**DATA ANALYSIS** 
<span style="font-size:xx-larger;"></span>





**1. Gender**
<span style="font-size:xx-larger;"></span>

```python
# Analyse distribution of sales by Gender
Gender_cat= pd.Categorical(Diwali_sales["Gender"])
Gender_cat.describe()
```

```python
# Visualaise distribution of sales by Gender
ax = sns.countplot(x = 'Gender',data = Diwali_sales)

for bars in ax.containers:
    ax.bar_label(bars)
```

According to the graph female customers tend to buy more than male customers.There are more than double female buyers than male buyers.

**2. AGE**
<span style="font-size:xx-larger;"></span>


```python
# Calculate summary statistics for age
print(Diwali_sales['Age'].describe())
```

```python
# Visualize age distribution
sns.histplot(Diwali_sales['Age'], bins=20)
plt.title('Age Distribution of Customers')
plt.xlabel("Age")
plt.show()
```

According to the graph people in the age group between 26-35 are the most frequent buyers. 

**3. MARITAL STATUS**
<span style="font-size:XX-larger;"></span>


```python
# Analyze the distribution of customers by marital status
ax = sns.countplot(x='Marital_Status', data=Diwali_sales)
plt.title('Distribution of Customers by Marital Status')

# Add count labels on each bar
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

According to the graph married people tend to buy more than unmarried people. 

```python
sales_state = Diwali_sales.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.title('Total Sales Amount by Marital Status and Gender')
```

According to the graph in married couples married females tend to purchase more than married male.

**3. OCCUPATION**
<span style="font-size:XX-larger;"></span>


```python
# Analyze the distribution of customers by occupation
plt.figure(figsize=(10,6))
sns.countplot(y='Occupation', data=Diwali_sales, order=Diwali_sales['Occupation'].value_counts().index)
plt.title('Distribution of Customers by Occupation')
plt.ylabel('Occupation')
plt.xlabel('Count')
plt.show()
```

According to the graph IT, Healtcare, Aviation, Banking, Govt and Hospitality employees are the majority of the customers.

**4. PRODUCT CATEGORY**
<span style="font-size:XX-larger;"></span>


```python
# Determine the most frequently purchased product categories
product_category_counts = Diwali_sales['Product_Category'].value_counts()
print(product_category_counts)

# Visualize product categories
plt.figure(figsize=(10,6))
product_category_counts.plot(kind='bar', color='orange')
plt.title('Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

According to the graph Clothing & Apparel, food, Electronics & Gadgets, Footwear & shoes & Household items are the maximum bought product categories.

**GEOGRAPHICAL ANALYSIS**
<span style="font-size:XX-larger;"></span>


```python
# Analyze customer distribution by state
state_counts = Diwali_sales['State'].value_counts()
print(state_counts)

# Visualize customer distribution by state
plt.figure(figsize=(10,6))
state_counts.plot(kind='bar', color='green')
plt.title('Customer Distribution by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

According to the graph the maximum customers are from UP, Maharashtra, Kartnataka, Delhi & MP

```python
# Compare sales performance across different zones
plt.figure(figsize=(8,5))
sns.countplot(x='Zone', data=Diwali_sales, order=Diwali_sales['Zone'].value_counts().index)
plt.title('Sales Performance by Zone')
plt.xlabel('Zone')
plt.ylabel('Count')
plt.show()
```

According to the graph Central and Southern regions are where the maximum customers come from.
